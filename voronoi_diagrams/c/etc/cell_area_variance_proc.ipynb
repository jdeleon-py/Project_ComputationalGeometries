```python
get_svar = lambda arr, mean, arr_n: (1 / (arr_n - 1)) * np.sum([(i - mean) ** 2 for i in arr])
print(get_svar(proc_norm, np.mean(proc_norm), 1001))
print(get_svar(proc_uni, np.mean(proc_uni), 1001))
```

```python
# Combine into a list for plotting
data = [proc_norm, proc_uni]

# Labels for the groups
labels = ['Uniform Sites', 'Normal Sites']

# Create the boxplot
plt.figure(figsize=(6, 4))
plt.boxplot(data, labels=labels, patch_artist=True,
            boxprops=dict(facecolor='white', color='black'),
            medianprops=dict(color='black'),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'),
            flierprops=dict(marker='o', color='gray', alpha=0.5))

plt.ylabel("Cell Area Variance (sq. px.)")
plt.title("Voronoi Cell Area Variance by Site Distribution")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
