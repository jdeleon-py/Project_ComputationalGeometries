```python
# DIM = 512x512

vor_sites_dim = np.array([100, 200, 300, 400, 500, 1000, 2000, 3000, 4000, 5000, 10000])

mem_qt_dim_avg = np.array([8.11e5, 1.13e6, 1.34e6, 1.52e6, 1.64e6, 2.14e6, 2.70e6, 3.03e6, 3.26e6, 3.43e6, 3.89e6])
time_qt_dim_avg = np.array([0.0852, 0.230, 0.357, 0.489, 0.624, 1.56, 3.82, 6.29, 8.97, 11.8, 26.8])

mem_na_dim = np.array([1.26e7, 1.26e7, 1.26e7, 1.26e7, 1.26e7, 1.26e7, 1.26e7, 1.26e7, 1.26e7, 1.26e7, 1.26e7])
time_na_dim_avg = np.array([0.336, 0.592, 0.859, 1.13, 1.36, 2.58, 5.30, 7.93, 10.3, 12.8, 26.2])
```

```python
# VOR SITES = 500

vor_sites_vor = [64, 128, 256, 512, 1024]

mem_qt_vor_avg = np.array([6.52e4, 2.33e5, 6.65e5, 1.66e6, 3.75e6])
time_qt_vor_avg = np.array([0.0252, 0.100, 0.287, 0.612, 1.44])

mem_na_vor_avg = np.array([1.97e5, 7.86e5, 3.14e6, 1.26e7, 5.03e7])
time_na_vor_avg = np.array([0.0235, 0.0984, 0.382, 1.34, 5.36])
```

```python
# (no render) data plotting
```

```python
# generation time vs. number of voronoi sites
plt.title("QT Gen Time with Window DIM (512x512) px")
plt.xlabel("Voronoi Sites")
plt.ylabel("Generation Time (sec)")

plt.scatter(vor_sites_dim, time_qt_dim_avg)
plt.plot(vor_sites_dim, time_qt_dim_avg)

plt.scatter(vor_sites_dim, time_na_dim_avg)
plt.plot(vor_sites_dim, time_na_dim_avg)

plt.semilogx()
plt.show()
```

```python
# memory usage vs. number of voronoi sites
plt.title("QT Memory with Window DIM (512x512) px")
plt.xlabel("Voronoi Sites")
plt.ylabel("Memory Usage (bytes)")

plt.scatter(vor_sites_dim, mem_qt_dim_avg)
plt.plot(vor_sites_dim, mem_qt_dim_avg)

plt.scatter(vor_sites_dim, mem_na_dim)
plt.plot(vor_sites_dim, mem_na_dim)

plt.semilogx()
plt.show()
```

```python
# generation time vs. window size
plt.title("QT Gen Time with 500 Voronoi Sites")
plt.xlabel("Window DIM (DIMxDIM) px")
plt.ylabel("Generation Time (sec)")

plt.scatter(vor_sites_vor, time_qt_vor_avg)
plt.plot(vor_sites_vor, time_qt_vor_avg)

plt.scatter(vor_sites_vor, time_na_vor_avg)
plt.plot(vor_sites_vor, time_na_vor_avg)

#plt.semilogx()
plt.show()
```

```python
# mem usage vs window size
plt.title("QT Mem Usage with 500 Voronoi Sites")
plt.xlabel("Window DIM (DIMxDIM) px")
plt.ylabel("Memory Usage (bytes)")

plt.scatter(vor_sites_vor, mem_qt_vor_avg)
plt.plot(vor_sites_vor, mem_qt_vor_avg)

plt.scatter(vor_sites_vor, mem_na_vor_avg)
plt.plot(vor_sites_vor, mem_na_vor_avg)

#plt.semilogx()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2*x)
y4 = np.cos(2*x)

# Create a figure and a grid of subplots (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot the first subplot (top-left)
axes[0, 0].plot(x, y1, color='blue', label='sin(x)')
axes[0, 0].set_title('Sin(x)')
axes[0, 0].legend()

# Plot the second subplot (top-right)
axes[0, 1].plot(x, y2, color='green', label='cos(x)')
axes[0, 1].set_title('Cos(x)')
axes[0, 1].legend()

# Plot the third subplot (bottom-left)
axes[1, 0].plot(x, y3, color='red', label='sin(2x)')
axes[1, 0].set_title('Sin(2x)')
axes[1, 0].legend()

# Plot the fourth subplot (bottom-right)
axes[1, 1].plot(x, y4, color='purple', label='cos(2x)')
axes[1, 1].set_title('Cos(2x)')
axes[1, 1].legend()

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```
